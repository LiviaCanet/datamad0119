```python
#Challenge 1
'''
You will find out how many programming languages are used among all the forks created from the main lab repo of your bootcamp. Assuming the main lab repo is ironhack-datalabs/madrid-oct-2018, you will:

Obtain the full list of forks created from the main lab repo via Github API.

Loop the JSON response to find out the language attribute of each fork. Use an array to store the language attributes of each fork.

Hint: Each language should appear only once in your array.
Print the language array. It should be something like:

["Python", "Jupyter Notebook", "HTML"]
'''

url = 'https://api.github.com/repos/ironhack-datalabs/datamad0119/forks'

response = requests.get(url)
results = response.json()
results
```

```python
#Para obtener language usando un loop

languages = []
for result in results:
    for ff in result:
        if ff == 'language':
            languages.append(result[ff])

print(set(languages))
    
```

```python
#Challenge 2
'''
Count how many commits were made in the past week.

Obtain all the commits made in the past week via API, which is a JSON array that contains multiple commit objects.

Count how many commit objects are contained in the array.

'''
url = 'https://api.github.com/repos/ironhack-datalabs/datamad0119/commits?since="2019-01-28T00:00:00Z"'
response = requests.get(url)
results = response.json()
results

```

```python
#Sacar los commits con un loop

counter = 0
for result in results:
    for x in result:
        if x == 'commit':
            counter += 1

counter
    
```

```python
#Solución Pablo:

response = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents')
results = response.json()

def scavengerhunt():
    data_js = []
    for result in results:
        if result["name"] == '.gitignore':
            pass
        else:
            datas = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{}'.format(result["name"]))
            data_js += datas.json()
    return data_js

data_js = scavengerhunt()

def dataaaaa():
    lista = {}
    for bo in data_js:
        if bo["name"].endswith('.scavengerhunt'):
            lista[bo["name"]] = bo["download_url"]
    return lista

lista = dataaaaa()

def ordedic():
    dic ={}
    for key in sorted(lista):
        dic[key] = lista[key]
    return dic

dic = ordedic()

def cadena():
    lista = []
    for key in dic:
        lista.append(requests.get(dic[key]).text[:-1])
    cadena = " ".join(lista)
    return cadena

cadena()
```
