```python
import pandas as pd
```

```python
data = pd.read_csv('../GSAF5.csv')
data
```

```python
#1-. Drop Duplicates rows

before = len(data)
data = data.drop_duplicates()
after = len(data)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
#2-. Drop Duplicates columns

select_columns = ['Case Number', 'Date', 'Year', 'Country', 'Area', 'Location', 'Activity', 'Name','Sex ', 'Species ', 'Investigator or Source', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22', 'Unnamed: 23']

data = data[select_columns].drop_duplicates()
after = len(data)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
#3-. Data Type

data.dtypes
# Podemos ver que hay un mix de tipos de data en todas las columnas menos en 'year' y en 'original order'. Tal y como tenemos los datos, sÃ³lo podrÃ­amos analizar estas dos variables.
```

```python
#4-. Renaming columns
#Renombro el nombre de las columnas porque tenÃ­an un error, un espacio de mÃ¡s.
data = data.rename(columns={'Sex ':'Sex',
                            'Species ':'Species'})
```

```python
#5-. Missing Values

null_cols = data.isnull().sum()
null_cols[null_cols > 0]
#Como podemos ver, casi el 100% de Unnamed 22 y Unnamed 23 estÃ¡n vacÃ­as, las podemos borrar. La columna Species tambiÃ©n estÃ¡ muy vacÃ­a, pero no merece la pena borrarla porque podemos sacar alguna conclusiÃ³n de los datos que hay.
```

```python
#6-. Dropping Useless Columns 

drop_cols = list(null_cols[null_cols >= 5990].index)
data = data.drop(drop_cols, axis=1)
data
```

```python
#TambiÃ©n borramos href, porque es prÃ¡cticamente igual que href formula, pero mÃ¡s completa.
data = data.drop('href', 1)
data
```

```python
#TambiÃ©n vamos a borrar Case Number2 porque, aunque no son duplicadas, podemos ver que prÃ¡cticamente son iguales. 
data = data.drop('Case Number.2', 1)
data
```

```python

#Un 2% de los aÃ±os estÃ¡n a 0 porque no se sabe el date exacto. Vamos a eliminar estos datos porque no nos deja hacer estadÃ­sticas y perderlos sÃ³lo supone perder un 2%, que es poco
#TambiÃ©n eliminamos los aÃ±os menores al 1900, el anÃ¡lisis serÃ¡ mÃ¡s actual.
data_year = data[data['Year'] < 1900]
data = data.drop(data_year.index)
data


```

```python
#Eliminamos la columna Date, porque no nos aporta informaciÃ³n en los primeros aÃ±os.
data = data.drop('Date', 1)
data
```

```python
#Eliminamos la columna Location, porque no nos aporta informaciÃ³n del todo (ya tenemos Area) y ademÃ¡s faltan valores.
data = data.drop('Location', 1)
data
```

```python
#Vamos a filtrar por el paÃ­s y el aÃ±o donde han ocurrido mÃ¡s ataques de tiburones. 

test = data[(data['Country']=='USA') & (data['Year']==2015)]
test

#Tenemos 72 casos de ataques de tiburones en EEUU el aÃ±o 2015.
```

```python
#Vamos a ver la frecuencia de accidentes en cada Ãrea, ordenado por PaÃ­s.

frecuencia_area = data.groupby(['Country','Area']).size()
frecuencia_area
```

```python
#De la tabla anterior, vamos a filtrar por las Ã¡reas donde ha habido mÃ¡s de 10 accidentes, donde podemos considerar que es un Ã¡rea peligrosa.

area_peligrosa = frecuencia_area.drop(frecuencia_area[frecuencia_area < 10].index)
area_peligrosa
```

```python
#Ahora tenemos la lista de las Ã¡reas donde tienes mÃ¡s riesgo de ser atacado por un tiburÃ³n, ordenadas de mÃ¡s a menos peligrosa.

area_peligrosa.sort_values(ascending=False)
```
