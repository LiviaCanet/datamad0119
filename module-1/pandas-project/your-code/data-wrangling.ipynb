```python
import pandas as pd
```

```python
data = pd.read_csv('../GSAF5.csv')
data
```

```python
#1-. Drop Duplicates rows

before = len(data)
data = data.drop_duplicates()
after = len(data)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
#2-. Drop Duplicates columns

select_columns = ['Case Number', 'Date', 'Year', 'Country', 'Area', 'Location', 'Activity', 'Name','Sex ', 'Species ', 'Investigator or Source', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22', 'Unnamed: 23']

data = data[select_columns].drop_duplicates()
after = len(data)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
#3-. Data Type

data.dtypes
# Podemos ver que hay un mix de tipos de data en todas las columnas menos en 'year' y en 'original order'. Tal y como tenemos los datos, sÃ³lo podrÃ­amos analizar estas dos variables.
```

```python
#4-. Renaming columns
#Renombro el nombre de las columnas porque tenÃ­an un error, un espacio de mÃ¡s.
data = data.rename(columns={'Sex ':'Sex',
                            'Species ':'Species'})
```

```python
#5-. Missing Values

null_cols = data.isnull().sum()
null_cols[null_cols > 0]
#Como podemos ver, casi el 100% de Unnamed 22 y Unnamed 23 estÃ¡n vacÃ­as, las podemos borrar. La columna Species tambiÃ©n estÃ¡ muy vacÃ­a, pero no merece la pena borrarla porque podemos sacar alguna conclusiÃ³n de los datos que hay.
```

```python
#6-. Dropping Useless Columns 

drop_cols = list(null_cols[null_cols >= 5990].index)
data = data.drop(drop_cols, axis=1)
data
```

```python
#TambiÃ©n borramos href, porque es prÃ¡cticamente igual que href formula, pero mÃ¡s completa.
data = data.drop('href', 1)
data
```

```python
#TambiÃ©n vamos a borrar Case Number2 porque, aunque no son duplicadas, podemos ver que prÃ¡cticamente son iguales. 
data = data.drop('Case Number.2', 1)
data
```
