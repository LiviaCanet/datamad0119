```python
import pandas as pd
```

```python
data = pd.read_csv('../GSAF5.csv')
data
```

```python
#1-. Drop Duplicates rows

before = len(data)
data = data.drop_duplicates()
after = len(data)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
#2-. Drop Duplicates columns

select_columns = ['Case Number', 'Date', 'Year', 'Country', 'Area', 'Location', 'Activity', 'Name','Sex ', 'Species ', 'Investigator or Source', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22', 'Unnamed: 23']

data = data[select_columns].drop_duplicates()
after = len(data)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
#3-. Data Type

data.dtypes
# Podemos ver que hay un mix de tipos de data en todas las columnas menos en 'year' y en 'original order'. Tal y como tenemos los datos, sólo podríamos analizar estas dos variables.
```

```python
#4-. Renaming columns
#Renombro el nombre de las columnas porque tenían un error, un espacio de más.
data = data.rename(columns={'Sex ':'Sex',
                            'Species ':'Species'})
```

```python
#5-. Missing Values

null_cols = data.isnull().sum()
null_cols[null_cols > 0]
#Como podemos ver, casi el 100% de Unnamed 22 y Unnamed 23 están vacías, las podemos borrar. La columna Species también está muy vacía, pero no merece la pena borrarla porque podemos sacar alguna conclusión de los datos que hay.
```

```python
#6-. Dropping Useless Columns 

drop_cols = list(null_cols[null_cols >= 5990].index)
data = data.drop(drop_cols, axis=1)
data
```

```python
#También borramos href, porque es prácticamente igual que href formula, pero más completa.
data = data.drop('href', 1)
data
```

```python
#También vamos a borrar Case Number2 porque, aunque no son duplicadas, podemos ver que prácticamente son iguales. 
data = data.drop('Case Number.2', 1)
data
```

```python

#Un 2% de los años están a 0 porque no se sabe el date exacto. Vamos a eliminar estos datos porque no nos deja hacer estadísticas y perderlos sólo supone perder un 2%, que es poco
#También eliminamos los años menores al 1900, el análisis será más actual.
data_year = data[data['Year'] < 1900]
data = data.drop(data_year.index)
data


```

```python
#Eliminamos la columna Date, porque no nos aporta información en los primeros años.
data = data.drop('Date', 1)
data
```

```python
#Eliminamos la columna Location, porque no nos aporta información del todo (ya tenemos Area) y además faltan valores.
data = data.drop('Location', 1)
data
```

```python
#Vamos a ver la frecuencia de accidentes en cada Área, ordenado por País.

frecuencia_area = data.groupby(['Country','Area']).size()
frecuencia_area
```

```python
#De la tabla anterior, vamos a filtrar por las áreas donde ha habido más de 10 accidentes, donde podemos considerar que es un área peligrosa.

area_peligrosa = frecuencia_area.drop(frecuencia_area[frecuencia_area < 10].index)
area_peligrosa
```

```python
#Ahora tenemos la lista de las áreas donde tienes más riesgo de ser atacado por un tiburón, ordenadas de más a menos peligrosa.

area_peligrosa.sort_values(ascending=False)
```
