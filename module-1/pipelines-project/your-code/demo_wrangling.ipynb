```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Adquire

demo = pd.read_csv('country_population.csv')
```

```python
demo.head(11)
```

```python
#Missing Values

null_cols = demo.isnull().sum()
null_cols[null_cols > 0]
```

```python
# Eliminamos las columnas que no vamos a utilizar

dropyears = []
for x in range(1960, 2000):
    dropyears.append(str(x))
    
print(dropyears)
    
```

```python
# Eliminamos las columnas que no vamos a utilizar

demo = demo.drop(dropyears, axis=1)
demo.head(11)
```

```python
# Eliminamos las columnas que no vamos a utilizar

demo = demo.drop(['Indicator Name', 'Indicator Code', 'Country Code'], axis=1)
demo.head(11)
```

```python
# Eliminamos las filas que tienen el valor nulo

droprows = []
for x in range(2000, 2017):
    droprows.append(str(x))
print(droprows)
```

```python
# Eliminamos las filas que tienen el valor nulo
demo = demo.dropna(subset = droprows, axis=0)
demo.head(11)
```

```python
#De aquí en adelante, calculate percentage change: pct_change()

demo_series = demo.apply(lambda x: x.tolist(), axis=1)
demo_series
```

```python
#Tenemos que transponer el Dataframe para poder recorrer cada serie.

demoT = demo.T
demoT.columns = demoT.iloc[0]
demoT.head()
```

```python
#Eliminamos la fila que hemos puesto como título

demoT = demoT.drop('Country Name', axis=0)
demoT
```

```python
#Ahora, podemos aplicar el pct_change() porque cada columna es una series.

print(type(demoT))
print(type(demoT.Aruba))
```
