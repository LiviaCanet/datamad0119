```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Adquire

demo = pd.read_csv('country_population.csv')
```

```python
demo.head(11)
```

```python
#Missing Values

null_cols = demo.isnull().sum()
null_cols[null_cols > 0]
```

```python
# Eliminamos las columnas que no vamos a utilizar

dropyears = []
for x in range(1960, 1970):
    dropyears.append(str(x))
    
print(dropyears)
    
```

```python
# Eliminamos las columnas que no vamos a utilizar

demo = demo.drop(dropyears, axis=1)
demo.head(11)
```

```python
# Eliminamos las columnas que no vamos a utilizar

demo = demo.drop(['Indicator Name', 'Indicator Code', 'Country Code'], axis=1)
demo.head(11)
```

```python
# Eliminamos las filas que tienen el valor nulo

droprows = []
for x in range(1970, 2017):
    droprows.append(str(x))
print(droprows)
```

```python
# Eliminamos las filas que tienen el valor nulo
demo = demo.dropna(subset = droprows, axis=0)
demo.head(11)
```

```python
#De aquí en adelante, calculate percentage change: pct_change()

demo_series = demo.apply(lambda x: x.tolist(), axis=1)
demo_series
```

```python
#Tenemos que transponer el Dataframe para poder recorrer cada serie.
#.iloc para poner los años como index.
demoT = demo.T
demoT.columns = demoT.iloc[0]
demoT
```

```python
#Eliminamos la fila que hemos puesto como título

demoT = demoT.drop('Country Name', axis=0)
demoT
```

```python
#Vamos a crear una tabla con los 5 países más poblados. 
#Queremos comparar su crecimiento de población.
demoT_maspob = demoT[['China', 'India', 'United States', 'Indonesia', 'Pakistan']].copy()
demoT_maspob
```

```python
#Convertir los index en ints para que lo coja el matplotlib. Intento fallido de usar matplotlib.
demoT_maspob_index = demoT_maspob.index.astype(int)
```

```python
print(type(demoT_maspob_index[0]))

#Vemos que ahora los índices son ints, debería cogerlo el matplotlib
```

```python
lines = demoT_maspob.plot.line(title = "Most populated countries \n Population \n x axis = 70's, 80's, 90's, 00's, 10's", legend=True, use_index=True, figsize=(16,9))
lines

#Podemos ver cómo China en los 80's alcanza el billón de habitantes y sigue creciendo hasta los 00's que frena un poco por el single child policy.
#Igualmente India llega al billón más adelante, en los 00's, y sigue creciendo hasta alcanzar prácticamente a China hoy en día.
```

```python
#Podemos aplicar el pct_change() porque cada columna es una series.

print(type(demoT))
print(type(demoT.Aruba))
```

```python
#La tabla de pct_change() de todos los países.

demoT_pctchange = demoT.pct_change()*100
demoT_pctchange = demoT_pctchange.round(2)
demoT_pctchange
```

```python
#Vamos a quedarnos con los 3 países más poblados y los 3 países menos poblados. 
#Queremos comparar su crecimiento de población con el de España y el total de crecimiento de población mundial.

demoT_pctchange_pob = demoT_pctchange[['China', 'India', 'United States', 'Spain', 'Tuvalu', 'Nauru', 'San Marino', 'World']].copy()
demoT_pctchange_pob


```

```python
lines = demoT_pctchange_pob.plot.line(title = "Percentage Change \n Most to least populated contries vs World population \n x axis = 70's, 80's, 90's, 00's, 10's \n y axis = %", legend=True, use_index=True, figsize=(17,10))
lines

#Vemos en esta tabla de Percentage Change cómo curiosamente Nauru, que es un estado de Micronesia, situado en el
#océano Pacífico, que comprende una sola isla justo al sur de la línea del ecuador; cambia drásticamente de población.
#En el 2014, Nauru aumentó su poplación en casi un 10% . Según datos publicados por la ONU, esto se debió a entrada
#de inmigrantes (refugiados). 
#También es interesante fijarse en la línea roja, España, vemos como en los años 80's la población disminuye. En los
#00's sube mucho, hasta un 2%.
```
