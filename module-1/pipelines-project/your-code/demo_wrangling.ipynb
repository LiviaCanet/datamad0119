```python
import pandas as pd
```

```python
#Adquire

demo = pd.read_csv('acs2015_county_data.csv')
```

```python
demo.head(11)
```

```python
#Drop Duplicates rows

before = len(demo)
demo = demo.drop_duplicates()
after = len(demo)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
#Missing Values

null_cols = demo.isnull().sum()
null_cols[null_cols > 0]
```

```python
# Eliminamos las columnas que no vamos a utilizar
demo = demo.drop(['IncomeErr', 'IncomePerCap', 'IncomePerCapErr', 'ChildPoverty', 'Drive', 'Carpool', 'Transit', 'Walk', 'OtherTransp', 'WorkAtHome', 'MeanCommute', 'PrivateWork', 'PublicWork', 'SelfEmployed', 'FamilyWork', 'Citizen'], axis=1)
demo.head(11)
```
