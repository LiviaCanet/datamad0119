####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
data = get_html(url)
````

````python
datetime = [re.sub("earthquake", "", e.text) for e in data.find_all("b")] #Tag b
datetime_cols = [e.split("\xa0\xa0\xa0") for e in datetime if "\xa0" in e]
latlong_cols = [e.text.strip() for e in data.find_all("td",{"class":"tabev1"})] #Tag td
latlong_cols = [latlong_cols[i:(i+2)] for i in range(0, (len(latlong_cols)), 2)] 
region_cols = [e.text.strip() for e in data.find_all("td",{"class":"tb_region"})]
earthquakes = pd.DataFrame(datetime_cols, columns = {"Date","Time"})
earthquakes[["Latitude","Longitude"]] = pd.DataFrame(latlong_cols)
earthquakes["Region"] = region_cols
earthquakes
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
#your code

data = get_html(url)
````

````python
table = data.find_all('table',{'class':'wikitable sortable'})
rows = [e.find_all('tr') for e in table][0]
rows = [row.text.strip().split("\n") for row in rows]
pd.DataFrame([[rows[i][1],rows[i][2]] for i in range(1,len(rows))], columns = {"Language", "Native speakers"})
````
