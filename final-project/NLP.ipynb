```python
s = '07/04/2010: On Sunday, in Andar, Ghazni, Afghanistan, suspected Taliban militants kidnapped six employees of Faizullah Construction Company. The six were on their way from work when they were abducted. On 07/05/2010, the militants shot and killed the hostages, then dumped the bodies of the six employees in the Andar district and Arzo village. No damages were reported and no group claimed responsibility.'
```

```python
def clean_up(s):
    s = re.sub("\d{2}/\d{2}/\d{4}:", ' ', s) #clean dates
    s = re.sub('\d+', ' ', s) # /d Any numeric character
    s = re.sub('\W+', ' ', s) # Any non-alphanumeric character
    return s.lower().strip()

clean_up(s)
```

```python
def tokenize(s):
    return word_tokenize(clean_up(s))
```

```python
def stem_and_lemmatize(s):
    return [WordNetLemmatizer().lemmatize(SnowballStemmer("english").stem(x)) for x in tokenize(s)]
```

```python
def remove_stopwords(s):
    return [w for w in stem_and_lemmatize(s) if not w in stopwords.words('english')]

remove_stopwords(s)
```

```python
def find_positive_features(document):
    words = set(document)
    features = {}
    for w in top:
        features[w] = (w in words)
    s = SentimentIntensityAnalyzer().polarity_scores(" ".join(document))

    #return (features, s["pos"], s["neg"])
    return (s["pos"])

feature = list(gt_10.news_processed.apply(find_positive_features))
```
